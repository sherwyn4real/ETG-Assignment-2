# Summer Olympics Data Analysis Assignment

```python
import pandas as pd

df = pd.read_csv('summer.csv')
#df.isnull().sum()      #This returned 4 in the column of Country
```

### 1. In how many cities Summer Olympics is held so far?

```python
print(len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
newdf = df[df['Medal'] == 'Gold']     #newdf is the dataframe having only Gold medal rows

l=[]
for sport in newdf['Sport'].unique():
    l.append([sport,len(newdf[newdf['Sport'] == sport])])
    
pd.DataFrame(l,columns=['sport','Gold medals']).sort_values(by='Gold medals',ascending=False).head().plot(x='sport',y='Gold medals',kind='bar',figsize=(8,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
l=[]

for sport in df['Sport'].unique():
    l.append([sport,len(df[df['Sport'] == sport])])

#l = pd.DataFrame(l,columns = ['sport','medals'])               Converts to DataFrame
#l.sort_values(by='medals',ascending=False,inplace=True)        Sorts sports based on medal count 

pd.DataFrame(l,columns=['sport','medals']).sort_values(by='medals',ascending=False).head().plot(x='sport',y='medals',kind='bar',figsize=(8,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
l=[]
for athlete in df['Athlete'].unique():
    l.append([athlete,len(df[df['Athlete'] == athlete])])


pd.DataFrame(l,columns=['athlete','medals']).sort_values(by='medals',ascending=False).head().plot(x='athlete',y='medals',kind='bar',figsize=(8,5))
```

### 5. Which player has won most number of Gold Medals? (Top 5)

```python
l = []
a = df[df['Medal'] == 'Gold']       #a is the dataframe having only those rows that have Gold

for ath in a['Athlete'].unique():
    l.append([ath,len(a[a['Athlete'] == ath] )])
    
pd.DataFrame(l,columns=['athlete', 'medal']).sort_values(by='medal',ascending=False).head().plot(x='athlete',y='medal',kind='bar',figsize=(8,5))

```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
ind = df[df['Country'] == 'IND']     #ind is the dataframe having only india
ind_gold = ind[ind['Medal'] == 'Gold']    #ind_gold is the dataframe having only gold medals won by india

g = []

for year in ind_gold['Year'].unique():
    g.append(year)
    
g.sort()
print("India won its first gold medal in the year:",g[0])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
e = []

for event in df['Event'].unique():
    e.append([event, len(df[df['Event'] == event])])
    
pd.DataFrame(e,columns = ['event','athletes']).sort_values(by='athletes',ascending=False).head().plot(x='event',y='athletes',kind='bar',figsize=(8,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
wom = df[df['Gender'] == 'Women']   #wom is the dataframe having only women
data = []
wom_gold = wom[wom['Medal'] == 'Gold']    #wom_gold is the dataframe having only women gold medalists

for sport in wom_gold['Sport'].unique():
    data.append([sport,len(wom_gold[wom_gold['Sport'] == sport])])
    
pd.DataFrame(data,columns=['sport','medals']).sort_values(by='medals', ascending = False).head().plot(x='sport',y='medals',kind='bar',figsize=(8,4))
```
